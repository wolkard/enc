```python
def get_binary_graph(img_g):
    img=cv2.adaptiveThreshold(img_g,255,cv2.ADAPTIVE_THRESH_MEAN_C ,cv2.THRESH_BINARY_INV,51,25)
    kernel= np.ones((5,5),np.uint8)
    img = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
    img = 255-img
    return img
```

```python
# remove side  and  add side
def remove_and_add_side(img_g):
    img_index = np.where(img_g<150)
    y = max(img_index[0])-min(img_index[0])
    x = max(img_index[1])-min(img_index[1])
    
    
    left = random.randint(1,20)
    right = random.randint(1,20)
    top = random.randint(1,10)
    bottom = random.randint(1,10)
    new_img = np.ones((y+top+bottom,x+left+right))*255
    print(top,bottom,left,right)
    new_img[top:-bottom,left:-right] = img_g[min(img_index[0]):max(img_index[0]),min(img_index[1]):max(img_index[1])]
    return new_img
```

```python
plt.rcParams['figure.figsize'] = (15.0, 8.0)
path = "../data/add_train/Img/Sample011/img011-004.png"
img = cv2.imread(path,0)
binary_grap_img = get_binary_graph(img)
remove_and_add_side_img = remove_and_add_side(binary_grap_img)
# plt.imshow(img)
# plt.show()
# plt.imshow(binary_grap_img)
# plt.show()
plt.imshow(remove_and_add_side_img)
plt.show()
```
